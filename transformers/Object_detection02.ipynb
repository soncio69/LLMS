```python
import cv2

stream = cv2.VideoCapture(0)  # Default webcam 
while(True):
    (grabbed, frame) = stream.read()    # Captures frame by frame 
    cv2.imshow("Image", frame)  # Shows the frame 
    key = cv2.waitKey(1) & 0xFF  
    if key == ord("q"):  # Press q to break out of the loop. 
        break

# Cleanup 
stream.release() 
cv2.waitKey(1)  
cv2.destroyAllWindows() 
cv2.waitKey(1) 
```

```python
from transformers import pipeline
from PIL import Image
import cv2
font   = cv2.FONT_HERSHEY_SIMPLEX  # Font, color, and line thickness for drawing rectangles and text 
color  = (0, 255, 255)   #1
stroke = 2               #1
detection = pipeline("object-detection", model="facebook/detr-resnet-50")   # Loads the object detection model 
stream = cv2.VideoCapture(0)  # Default webcam 
while(True):  
    (grabbed, frame) = stream.read()  # Captures frame by frame 
    image = Image.fromarray(frame)    # Converts the video image from NumPy array to a PIL image 
    results = detection(image)  # Detects objects in the image 
    for object in results:      
        box = [i for i in object['box'].values()]       # Gets the coordinates for the object detected 
        # Draws bounding box around object detected 
        cv2.rectangle(frame, 
                      (box[0],box[1]),   
                      (box[2],box[3]),   
                      color, stroke)        
        # Draws the label for the object 
        cv2.putText(frame, f'({object["label"]})',
                    (box[0],box[1]-8),  
                    font, 1, color,  
                    stroke, cv2.LINE_AA)    
    cv2.imshow("Image", frame)  # Shows the frame 
    key = cv2.waitKey(1) & 0xFF  
    if key == ord("q"):  # Press q to break out of the loop. 
        break
# Cleanup 
stream.release()  
cv2.waitKey(1)  
cv2.destroyAllWindows()
cv2.waitKey(1) 
```
